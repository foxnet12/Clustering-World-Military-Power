```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("global firepower 2022 wide.csv")
```

## Select the relevant features for clustering

```python
features = ['Active Personnel', 'Armored Vehicles', 'Attack Helicopters', 'Defense Budget', 'Fighters/Interceptors', 'Total Population']
```

## Preprocess the data by standardizing the features

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df[features])
```

## Perform clustering using K-means algorithm

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_df)
```

## Add the cluster labels to the original data

```python
df['Cluster'] = kmeans.labels_
```

##  Visualize the clusters

```python
plt.scatter(df['Defense Budget'], df['Total Population'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Defense Budget')
plt.ylabel('Total Population')
plt.title('Clustering of Countries based on Defense Budget and Total Population')
plt.show()
```
